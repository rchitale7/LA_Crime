# Crime in West Los Angeles

Data analysis and visualization of crimes that have happened in the past 10 years in West Los Angeles. All data is from Los Angeles Police Department. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Crime_Data.csv') #crime from LAPD
```

We'll first look at crime across all sub-areas of Los Angeles:

```python
dfAll = pd.read_csv('Crime_Data.csv')
allCounts = dfAll["Area Name"].value_counts()
print(allCounts)
allCounts.plot.bar()
```

As shown by the bar plot above, West LA is on the lower end of this crime spectrum. Now, let's examine crimes that happen in West LA further. We are going to add columns for year, month, and date to existing data for future use, and we'll filter the data so we only see crime from West LA.  

```python
df = df[df['Area Name'] == "West LA"]
```

```python
df["YR"] = [s[6:] for s in df["Date Occurred"]]
df["M"] = [s[0:2] for s in df["Date Occurred"]]
df["Day"] = [s[3:5] for s in df["Date Occurred"]]
```

```python
df["YMD"] = df.apply(lambda x: x.YR + x.M + x.Day, axis = 1)
df["YMD"] = df["YMD"].astype(int) #YR/M/D in numerical form
df
```

# Data Visualization

We'll first look at all the types of crimes have that happened over the past 10 years in West LA, using a pie chart. 

```python
countsCrimes = df["Crime Code Description"].value_counts()


countsCrimes.plot.pie(figsize=(50, 50), labels =None, title = "Crime Code Description", fontsize = 80)
plt.legend(labels=countsCrimes.index, loc = 4, fontsize  = 15)
```

Since so many different types of crimes have happened over the past 10 years, we'll only examine crimes that happen frequently in order to extract relevant information. From the pie chart above, it looks like Burglary and Petty Theft are among the most common types of crimes, but let's get actual numbers for those crime statistics:  

```python
countsCrimes.mean() #get the mean amongst all types of crimes
```

```python
countsCrimes = countsCrimes[countsCrimes >= countsCrimes.mean()]
```

```python
countsCrimes.plot.barh()
```

In the bar chart above, we can see that there were about 8000 burglary and burglary from vehicle crimes in West LA over the past 10 years. We see some deadly crimes - like Assault with Deadly Weapon, which has about 1500 crimes - , but most of these crimes are not deadly. Let's compare it with a neighborhood that has had a lot of crimes - 77th street:

```python
df77 = dfAll[dfAll["Area Name"] == "77th Street"]
counts77 = df77["Crime Code Description"].value_counts()
counts77 = counts77[counts77 >= counts77.mean()]
counts77.plot.barh()
```

We can see right away that some of the most frequent crimes in 77th street are violent - such as Battery and Assault with Deadly Weapon - unlike West LA. Let's go back to examining West LA crime rates. We'll look at crime rates by age range. 

```python
df["Age Range"] = df["Victim Age"]
length = df.shape[0]
for i in range(0, length):
    if df.iloc[i, 30] <= 18:
        df.iat[i, 30] = 0
    elif df.iloc[i, 30] <= 30:
        df.iat[i, 30] = 19
    elif df.iloc[i, 30] <= 50:
        df.iat[i, 30] = 31
    else:
        df.iat[i, 30] = 50


    
        
```

We can see that most victims of crimes are over the age of 30. Let's look at the specific types of crimes that happen to people in these age ranges. 

```python
df2 = df[df["Age Range"] == 31]
countsCrimes31 = df2["Crime Code Description"].value_counts()
countsCrimes31 = countsCrimes31[countsCrimes31 >= countsCrimes31.mean()]
countsCrimes31.plot.barh()
```

```python
df3 = df[df["Age Range"] == 50]
countsCrimes50 = df3["Crime Code Description"].value_counts()
countsCrimes50 = countsCrimes50[countsCrimes50 >= countsCrimes50.mean()]
countsCrimes50.plot.barh()
```

```python
df4 = df[df["Age Range"] == 19]
countsCrimes19 = df4["Crime Code Description"].value_counts()
countsCrimes19 = countsCrimes19[countsCrimes19 >= countsCrimes19.mean()]
countsCrimes19.plot.barh()
```

```python
df5 = df[df["Age Range"] == 0]
countsCrimes0 = df5["Crime Code Description"].value_counts()
countsCrimes0 = countsCrimes0[countsCrimes0 >= countsCrimes0.mean()]
countsCrimes0.plot.barh()
```

Burglary tends to be the most frequently committed crime amongst all range victims except for the 0-18 age range. The fact that burglary is the most common crime for most age ranges makes sense, given overall crimes in West LA. However, we can see this doesn't always apply to teenage victims, where stolen vehicles is by far the most frequent crime. Possible explanation: many of these teenage victims are not as responsible with their new cars and driving privileges. Finally, let's look at crimes by year:

```python
yearCounts = df["YR"].value_counts()
yearCounts = yearCounts.sort_index()

plt.xticks(np.arange(9), ('2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018'))
yearCounts.plot()
```

Crime in WestLA has noticeably fallen in the past 2 years, with 2015 having the highest level over this period. Let's look at the common types of crimes that happened in 2015: 

```python
df6 = df[df["YR"]=="2015"]
countsdf6 = df6["Crime Code Description"].value_counts()
```

```python
countsdf6 = countsdf6[countsdf6 >= countsdf6.mean()]
countsdf6.plot.barh()
```

We can see that burglary - the most common crime in West LA over the last 10 years - dominates here, but not as much as Identity Theft. Identity theft has also been a very common crime in West LA over the years, so this is plausible.
